```python
from math import sqrt
import itertools

def findHeuristic(cs, gs):
    
    #create function to calculate Manhattan distance 
    def manhattan(a, b):
        return sum(abs(val1-val2) for val1, val2 in zip(a,b))

    #define vectors
    A = cs
    B = gs
#     A = [[None, 1, 2], [3, 4, 5], [6, 7, 8]]
#     B = [[  7 , 2, 4], [ 5, None , 6], [ 8  , 3 , 1]]

    # flatennning the list of lists to a single list
    mergedA = list(itertools.chain.from_iterable(A))
    mergedB = list(itertools.chain.from_iterable(B))


    # initialize filtered list
    filtered_listA = []
    filtered_listB = []

    # Using for loop
    for eleA, eleB in zip(mergedA,mergedB):
        if(eleA != None and eleB != None):
            filtered_listA.append(eleA)
            filtered_listB.append(eleB)

        elif(eleA != None and eleB == None):
            filtered_listA.append(eleA)
            filtered_listB.append(0)

        elif(eleA == None and eleB != None):
            filtered_listA.append(0)
            filtered_listB.append(eleB)

        elif(eleA == None and eleB == None):
            filtered_listA.append(0)
            filtered_listB.append(0)


#     print(filtered_listA, filtered_listB, sep = "\n")
    
    gsHeu = manhattan(filtered_listA, filtered_listB)
    
#     code exit if goal stat is found ie., heu is zero
    if(gsHeu == 0):
        solution = copy.deepcopy(cs)
        print("kudos, goal reached!")
        print(cs)
        sys.exit()
        
    global minHeu, minHeuState
    if(minHeu >= gsHeu):
        minHeu = gsHeu
        minHeuState = mergedA
#         print(f"minimum heuristic is {minHeu} - for state is {minHeuState}")
        
    #calculate Manhattan distance between vectors
    return manhattan(filtered_listA, filtered_listB)
```
